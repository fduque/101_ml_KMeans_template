```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```
#creating dataset
#create some aleatory data
x = []
y = []
for i in range(1,1000):
  x.append(random.randrange(11, 51))
  y.append(random.randrange(11, 51))
print(x,y)
```

```
#converting into a dataFrame
list_zipped = list(zip(x,y))
df = pd.DataFrame(data=list_zipped, columns=["x", "y"])
df.head(5)
```

```
#kmeans lib
from sklearn.cluster import KMeans
```

```
#create the main object for cluster Algo
#setting initially K=2 
kmeans = KMeans(n_clusters=2)
#applying Algo
kmeans.fit(df)
#finding Centroids coordinates
centroids = kmeans.cluster_centers_
print(centroids)
```

```
#plotiing centroids
plt.scatter(df['x'],df['y'], c=kmeans.labels_.astype(float), s=50, alpha=0.5)
#marking centroids dots with red color
plt.scatter(centroids[:,0], centroids[:,1], c='red', s=90)
plt.xlabel('x')
plt.ylabel('Y')
plt.title('Example K = 2')
plt.savefig('kmeans=2.png')
```
